#### XGBOOSTS experiment

```python
# import libraries
import pandas as pd
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('/home/h6x/git_projects/ornl-overdose-modeling-per-images/XGBoost/merged_data.csv')
```

```python
df.head()
```

```python
# sort the data by NOD
df = df.sort_values(by='NOD', ascending=False)
```

```python
df.head()
```

```python
# add new column to the data frame as a label and by the quantile of NOD add the label
df['label'] = pd.qcut(df['NOD'], q=[0,0.80,1], labels=['0', '1'])
```

```python
df.head()
```

```python
df_train = df.drop(['NOD'], axis=1)
```

```python
df_train.head()
```

```python
y = df_train['label']

# turn string labels into integers
y = y.astype('int')
```

```python
# get the first 100 columns as features
x_set_1 = df_train.iloc[:, 0:100]
x_set_2 = df_train.iloc[:, 100:200]
x_set_3 = df_train.iloc[:, 200:300]
x_set_4 = df_train.iloc[:, 300:400]
x_set_5 = df_train.iloc[:, 400:500]
x_set_6 = df_train.iloc[:, 500:600]
x_set_7 = df_train.iloc[:, 600:700]
x_set_8 = df_train.iloc[:, 700:800]
x_set_9 = df_train.iloc[:, 800:900]
x_set_10 = df_train.iloc[:, 900:1000]
x_set_11 = df_train.iloc[:, 1000:1100]
x_set_12 = df_train.iloc[:, 1100:1200]
x_set_13 = df_train.iloc[:, 1200:1300]
x_set_14 = df_train.iloc[:, 1300:1400]
x_set_15 = df_train.iloc[:, 1400:1500]
```

```python
# Train-test split
X1_train, X1_test, y_train, y_test = train_test_split(x_set_1, y, test_size=0.2, random_state=42)
X2_train, X2_test, _, _ = train_test_split(x_set_2, y, test_size=0.2, random_state=42)
X3_train, X3_test, _, _ = train_test_split(x_set_3, y, test_size=0.2, random_state=42)
X4_train, X4_test, _, _ = train_test_split(x_set_4, y, test_size=0.2, random_state=42)
X5_train, X5_test, _, _ = train_test_split(x_set_5, y, test_size=0.2, random_state=42)
X6_train, X6_test, _, _ = train_test_split(x_set_6, y, test_size=0.2, random_state=42)
X7_train, X7_test, _, _ = train_test_split(x_set_7, y, test_size=0.2, random_state=42)
X8_train, X8_test, _, _ = train_test_split(x_set_8, y, test_size=0.2, random_state=42)
X9_train, X9_test, _, _ = train_test_split(x_set_9, y, test_size=0.2, random_state=42)
X10_train, X10_test, _, _ = train_test_split(x_set_10, y, test_size=0.2, random_state=42)
X11_train, X11_test, _, _ = train_test_split(x_set_11, y, test_size=0.2, random_state=42)
X12_train, X12_test, _, _ = train_test_split(x_set_12, y, test_size=0.2, random_state=42)
X13_train, X13_test, _, _ = train_test_split(x_set_13, y, test_size=0.2, random_state=42)
X14_train, X14_test, _, _ = train_test_split(x_set_14, y, test_size=0.2, random_state=42)
X15_train, X15_test, _, _ = train_test_split(x_set_15, y, test_size=0.2, random_state=42)
```

```python
# Train individual models
model1 = XGBClassifier().fit(X1_train, y_train)
model2 = XGBClassifier().fit(X2_train, y_train)
model3 = XGBClassifier().fit(X3_train, y_train)
model4 = XGBClassifier().fit(X4_train, y_train)
model5 = XGBClassifier().fit(X5_train, y_train)
model6 = XGBClassifier().fit(X6_train, y_train)
model7 = XGBClassifier().fit(X7_train, y_train)
model8 = XGBClassifier().fit(X8_train, y_train)
model9 = XGBClassifier().fit(X9_train, y_train)
model10 = XGBClassifier().fit(X10_train, y_train)
model11 = XGBClassifier().fit(X11_train, y_train)
model12 = XGBClassifier().fit(X12_train, y_train)
model13 = XGBClassifier().fit(X13_train, y_train)
model14 = XGBClassifier().fit(X14_train, y_train)
model15 = XGBClassifier().fit(X15_train, y_train)
```

```python
# Predictions
pred1 = model1.predict_proba(X1_test)[:, 1]
pred2 = model2.predict_proba(X2_test)[:, 1]
pred3 = model3.predict_proba(X3_test)[:, 1]
pred4 = model4.predict_proba(X4_test)[:, 1]
pred5 = model5.predict_proba(X5_test)[:, 1]
pred6 = model6.predict_proba(X6_test)[:, 1]
pred7 = model7.predict_proba(X7_test)[:, 1]
pred8 = model8.predict_proba(X8_test)[:, 1]
pred9 = model9.predict_proba(X9_test)[:, 1]
pred10 = model10.predict_proba(X10_test)[:, 1]
pred11 = model11.predict_proba(X11_test)[:, 1]
pred12 = model12.predict_proba(X12_test)[:, 1]
pred13 = model13.predict_proba(X13_test)[:, 1]
pred14 = model14.predict_proba(X14_test)[:, 1]
pred15 = model15.predict_proba(X15_test)[:, 1]
```

```python
# Combine predictions into a new dataset
meta_X = pd.DataFrame({'pred1': pred1, 'pred2': pred2, 'pred3': pred3, 'pred4': pred4, 'pred5': pred5, 'pred6': pred6, 'pred7': pred7, 'pred8': pred8, 'pred9': pred9, 'pred10': pred10, 'pred11': pred11, 'pred12': pred12, 'pred13': pred13, 'pred14': pred14, 'pred15': pred15})
```

```python
# train a meta-model on the predictions
meta_model = LogisticRegression().fit(meta_X, y_test)
```

```python
# Evaluate meta-model
accuracy = meta_model.score(meta_X, y_test)
print(f"Meta-Model Accuracy: {accuracy:.2f}")
```

```python
# get the f1 scores for the meta-model each class
from sklearn.metrics import f1_score

y_pred = meta_model.predict(meta_X)
f1 = f1_score(y_test, y_pred)
print(f"Meta-Model F1 Score: {f1:.2f}")

# get the confusion matrix for the meta-model
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)

# get the classification report for the meta-model
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
# Accessing coefficients of the meta-model
coefficients = meta_model.coef_

# Intercept (bias term)
intercept = meta_model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```
